```python
import pandas as pd
data = {
    "Month": ['Jan','Feb','Mar','Apr','May','Jun'],
    "Sales": [10000, 12000, 15000, 13000, 17000, 16000],
    "Profit": [2000, 3000, 4000, 2500, 3500,3000]
    }

df = pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt  #1 LINE PLOT MONTHLY SALES

plt.figure(figsize=(8,5))
plt.plot(df['Month'],df['Sales'],color='blue',marker='o',linestyle='-',label='Sales')
plt.title('Sales Trends Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales ($)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
width = 0.35
plt.bar(df['Month'],df['Sales'],width=width,color='green',label='Sales')
plt.bar(df['Month'],df['Profit'],width=width,color='purple',label='Profit',bottom=df['Sales'])
plt.title('Sales Trends Over Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales ($)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# PIE CHART PROFIT VS MONTH


```python
from enum import auto
plt.figure(figsize=(12,5))
plt.pie(df['Profit'], labels=df['Month'], autopct='%1.2f%%',startangle=140, colors=plt.cm.Paired.colors)
plt.title('Profit by Month')
plt.show()
```

# Scatter plot


```python
plt.figure(figsize=(8,5))
plt.scatter(df['Sales'], df['Profit'],color='red',s = 100, edgecolors='black')
plt.title('Sales vs Profit Scatter Plot')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```

# **Histogram**

```python
plt.figure(figsize=(10,5))
plt.hist(df['Sales'], bins=5, color='yellow', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequecy')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(df['Profit'], vert=False, patch_artist=True, boxprops=dict(facecolor='red'))
plt.title('Sales Box Plot'),
plt.ylabel('Frequecy'),
plt.tight_layout(),
plt.show()
```

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month": ['Jan','Feb','Mar','Apr','May','Jun'],
    "Sales": [10000, 12000, 15000, 13000, 17000, 16000],
    "Profit": [2000, 3000, 4000, 2500, 3500,3000]
}

df = pd.DataFrame(data)

def genrate_plot(plot_type):
    fig = plt.figure(figsize=(10,5))

    if plot_type == 'Line Plot':
      plt.plot(df['Month'],df['Sales'],color='blue',marker='o',linestyle='--',label='Sales')
      plt.title('Sales Trends Over Months')
      plt.xlabel('Month')
      plt.ylabel('Sales($)')
      plt.grid(True)
      plt.legend()

    elif plot_type == 'stacked bar chart':
      fig.set_size_inches(10,6)
      width = 0.35
      plt.bar(df['Month'],df['Sales'],width=width,color='green',label='Sales')
      plt.bar(df['Month'],df['Profit'],width=width,color='purple',label='Profit',bottom=df['Sales'])
      plt.title('Sales and Profit Trends by Months')
      plt.xlabel('Month')
      plt.ylabel('Amount ($)')
      plt.grid(True)
      plt.legend()


    elif plot_type == 'Pie Chart':
      fig.set_size_inches(7,8)
      plt.pie(df['Profit'], labels=df['Month'], autopct='%1.2f%%',startangle=140, colors=plt.cm.Paired.colors)
      plt.title('Profit by Month')
      # Pie charts don't usually have grids, so removing grid=True
      # Pie charts don't usually have legends as labels are shown on the slices.
      # However, if legend is needed, it can be added with plt.legend() after plt.pie()

    elif plot_type =='Scatter Plot':
      plt.scatter(df['Sales'], df['Profit'],color='red',s = 100, edgecolors='black')
      plt.title('Sales vs Profit Scatter Plot'),
      plt.xlabel('Sales'),
      plt.ylabel('Profit'),
      plt.grid(True),
      # Scatter plots usually don't have legends unless different groups are plotted.

    elif plot_type == 'Histogram':
      plt.hist(df['Sales'], bins=5, color='yellow', edgecolor='black'),
      plt.title('Sales Distribution'),
      plt.xlabel('Sales'),
      plt.ylabel('Frequecy'),
      # Histograms don't typically have legends.

    elif plot_type == 'Boxplot':
      plt.boxplot(df['Profit'], vert=False, patch_artist=True, boxprops=dict(facecolor='red')),
      plt.title('Profit Box Plot'), # Corrected title from 'Sales Box Plot' to 'Profit Box Plot'
      plt.ylabel('Frequecy'), # Box plots typically have labels on the axis representing the data, not frequency
      # Box plots do not have legends.


    plt.tight_layout()
    return fig

# gradio ui


demo = gr.Interface(
    fn=genrate_plot,
    inputs=gr.Radio(['Line Plot','stacked bar chart','Pie Chart','Scatter Plot','Histogram','Boxplot']),
    outputs=gr.Plot(label="Sales data visualization"),
    title = 'sales & Profit visual insight',
    description="choose the type to visualize the data"
    ) # Removed the extra comma here

demo.launch()
```
